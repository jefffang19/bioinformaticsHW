# map all letters to lower case; i.e. Aâ€“>a, Câ€“>c

```python
for i in range(len(S)):
    if(S[i].isupper()):
        S[i] = S[i].lower()
```

```python
for i in S:
    print(i, end='')
```

# Order 1

```python
counting = {'a':0, 't':0, 'c':0, 'g':0}
```

```python
for i in S:
    counting[i] += 1
```

```python
counting
```

```python
len(S)
```

```python
prob = {'a':counting['a']/len(S),'t':counting['t']/len(S),'c':counting['c']/len(S),'g':counting['g']/len(S)}
```

```python
prob
```

```python
import math
```

```python
#get log base 2 probablity
for i in prob.keys():
    prob[i] = math.log2(prob[i])
    
prob
```

$log_{2}(P) = log_{2}(p_{1}*p_{2}*p_{3}*p_{4}*...*p_{n}) = log_{2}(p_{1}) + log_{2}(p_{2}) + log_{2}(p_{3}) + ... + log_{2}(p_{n})$

```python
probability = 1
for i in S:
    probability+=prob[i]
    
probability
```
