# Viterbi Algorithm

```python
def Viterbi(obs, states, init, trans, emis):
    #dynamic programming
    V = [] #V[i][j]=prob => prob of i_th observation is state j 
    path = [] #Path[i][j]=p => from (i-1)_th observation, taken path p to get to i_th observation state j
    
    v = {}
    tmp = []
    # v[0][j]
    for st in states:
        v[st] = (init[st]*emis[st][obs[0]])
        tmp.append(0)
    
    path.append(tmp) 
    V.append(v)
       
    
    # v[i][j] (i > 0)
    for ob in range(1,len(obs)):
        v = {}
        tmp_path = []
        for st in states:
            tmp_v = []
            for prev_st in states:
                tmp_v.append(trans[prev_st][st]*emis[st][obs[ob]]*V[ob-1][prev_st])
            
            v[st] = np.max(tmp_v)
            tmp_path.append(np.argmax(tmp_v))
                
        V.append(v)
        path.append(tmp_path)
        
    
    def dp_table():
        for i in obs:
            print('\t{}'.format(i),end='\t\t')
        print('\t(observation)')
        
        for i in states:
            print('{}\t'.format(i),end='')
            for j in range(len(obs)):
                print(V[j][i],end='\t')
            print()
        
        print('(states)')
    
    def print_path():
        rev_path = []
        for i in range(len(obs)-1, -1, -1):
            rev_path.append(np.argmax(list(V[i].values())))
        
        rev_path.reverse()
        for i in rev_path:
            print(states[i],end=',')
    
    dp_table()
    print_path()
        
```

```python
observation = ['n','c','d']
states = ['H','F']
initial = {'H':0.6, 'F':0.4}
transition = {'H':{'H':0.7, 'F':0.3}, 'F':{'H':0.4,'F':0.6}}
emission = {'H':{'n':0.5, 'c':0.4, 'd':0.1}, 'F':{'n':0.1, 'c':0.3, 'd':0.6}}
```

```python
Viterbi(observation, states, initial, transition, emission)
```

# Viterbi Algorithm (probabilty as $log_{2}$ version)

```python
def Viterbi_log2(obs, states, init, trans, emis, show_table):
    #all the probabilties are processed with log2
    
    #dynamic programming
    V = [] #V[i][j]=prob => prob of i_th observation is state j 
    path = [] #Path[i][j]=p => from (i-1)_th observation, taken path p to get to i_th observation state j
    
    v = {}
    tmp = []
    # v[0][j]
    for st in states:
        v[st] = (init[st]+emis[st][obs[0]])
        tmp.append(0)
    
    path.append(tmp) 
    V.append(v)
       
    
    # v[i][j] (i > 0)
    for ob in range(1,len(obs)):
        v = {}
        tmp_path = []
        for st in states:
            tmp_v = []
            for prev_st in states:
                tmp_v.append(trans[prev_st][st]+emis[st][obs[ob]]+V[ob-1][prev_st])
            
            v[st] = np.max(tmp_v)
            tmp_path.append(np.argmax(tmp_v))
                
        V.append(v)
        path.append(tmp_path)
        
    def dp_table():
        for i in obs:
            print('\t{}'.format(i),end='\t\t')
        print('\t(observation)')
        
        for i in states:
            print('{}\t'.format(i),end='')
            for j in range(len(obs)):
                print(V[j][i],end='\t')
            print()
        
        print('(states)')
    
    def print_path():
        rev_path = []
        for i in range(len(obs)-1, -1, -1):
            rev_path.append(np.argmax(list(V[i].values())))
        
        rev_path.reverse()
        for i in rev_path:
            print(states[i],end=',')
    
    if(show_table):
        dp_table()
        print_path()
        
    print(np.max(list(V[len(obs)-1].values())))
        
```

```python
observation = ['n','c','d']
states = ['H','F']
initial = {'H':math.log2(0.6), 'F':math.log2(0.4)}
transition = {'H':{'H':math.log2(0.7), 'F':math.log2(0.3)}, 'F':{'H':math.log2(0.4),'F':math.log2(0.6)}}
emission = {'H':{'n':math.log2(0.5), 'c':math.log2(0.4), 'd':math.log2(0.1)}, 'F':{'n':math.log2(0.1), 'c':math.log2(0.3), 'd':math.log2(0.6)}}
```

```python
Viterbi_log2(observation, states, initial, transition, emission, True)
```

# Now we Build our HMM with random probabilities

```python
class HMM:
    def __init__(self, obs):
        self.observation = obs
        self.states = ['0','1','2']
        self.initial = {'0':math.log2(0.4), '1':math.log2(0.4), '2':math.log2(0.2)}
        self.transition = {'0':{'0':math.log2(0.6), '1':math.log2(0.3), '2':math.log2(0.1)}, '1':{'0':math.log2(0.1), '1':math.log2(0.3), '2':math.log2(0.6)}, '2':{'0':math.log2(0.4), '1':math.log2(0.3), '2':math.log2(0.4)} }
        self.emission = {'0':{'a':math.log2(0.4), 't':math.log2(0.4), 'c':math.log2(0.1), 'g':math.log2(0.1)}, '1':{'a':math.log2(0.1), 't':math.log2(0.7), 'c':math.log2(0.1), 'g':math.log2(0.1)}, '2':{'a':math.log2(0.3), 't':math.log2(0.2), 'c':math.log2(0.2), 'g':math.log2(0.3)} }
    
    def fit(self, show_table = False):
        Viterbi_log2(self.observation, self.states, self.initial, self.transition, self.emission, show_table)
```
